```python
import numpy as np
rng = np.random.default_rng(0)
n, p = 200, 20
X = rng.standard_normal((n, p))
Y = rng.choice(['A','B'], size=n)
```

```python
from glmnet.lognet import LogNet
L = LogNet()
L.fit(X, Y)
```

```python
L.summary_
```

```python
import rpy2
%load_ext rpy2.ipython
%R -i X,Y
```

```python
%%R
library(glmnet)
y = as.integer(Y=='A')
G = glmnet(X, y, family='binomial')
print(G)
plot(G)
```
